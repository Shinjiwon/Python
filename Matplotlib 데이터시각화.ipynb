```python
# matplotlib 라이브러리 import
# from matplotli import pyplot as plt
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [110, 130, 120]) # plot(x, y) 기본라인스타일, 색상
plt.show()
```

```python
plt.plot(['Seoul', 'Paris', 'Seatle'], [30, 25, 55]) # 데이터

# 차트의 구성요소
plt.xlabel('City') # x축 이름
plt.ylabel('Response') # y축 이름
plt.title('Experiment Result') # 차트 제목

plt.show() # 보여주기
```

```python
plt.plot([1,2,3], [1,4,9]) # Data1
plt.plot([2,3,4], [5,6,7]) # Data2

# 구성요소
plt.xlabel('Sequence') # x축 이름
plt.ylabel('Timer(secs)') # y축 이름
plt.title('Experiment Result') # 제목
plt.legend(['Mouse', 'Cat']) # 범례

plt.show()
```

```python
y = [5,3,7,10,9,5,5,8]
x = range(len(y)) # range(8) → 0,1,2,3,4,5,6,7
plt.bar(x, y, width=0.7, color='steelblue')

plt.show()
```
